```python
train_df = pd.read_table('train.tsv')
test_df = pd.read_table('test.tsv')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()


for coloumn in train_df.columns:
  if train_df[coloumn].dtype == 'object':
    train_df[coloumn] =label_encoder.fit_transform(train_df[coloumn])



for coloumn in test_df.columns:
  if test_df[coloumn].dtype == 'object':
    test_df[coloumn] =label_encoder.fit_transform(test_df[coloumn])
```

```python
y = train_df['LABEL']
x = train_df.drop("LABEL",axis=1 )
```

```python
x.head()
```

```python
x_arr = x.values
y_arr = y.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(x_arr,y_arr,test_size=0.3,random_state=42)
```

```python
def evaluate(clf):
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_val)
    train_score = np.mean(y_pred_train == y_train)
    val_score = np.mean(y_pred_test == y_val)

    print("Train score:", round(train_score, 4))
    print("Test score:", round(val_score, 4))

    if train_score - val_score > 0.05:
        print("Overfitting warning!!!")
```

```python
clf2 = AdaBoostClassifier()
clf2.fit(X_train, y_train)

print("AdaBoost")
evaluate(clf2)
```

```python
clf3 = GradientBoostingClassifier()
clf3.fit(X_train, y_train)

print("Gradient Boosting")
evaluate(clf3)
```

```python
clf4 = RandomForestClassifier()
clf4.fit(X_train, y_train)

print("Random Forest")
evaluate(clf4)
```

```python
clf5 = KNeighborsClassifier()
clf5.fit(X_train, y_train)

print("K Neighbors")
evaluate(clf5)
```

```python
clf6 = DecisionTreeClassifier()
clf6.fit(X_train, y_train)

print("Decision Tree")
evaluate(clf6)
```

```python
clf7 = ExtraTreesClassifier()
clf7.fit(X_train, y_train)

print("Extra Trees")
evaluate(clf7)
```

```python
clf8 = SVC(probability=True, gamma='auto')
clf8.fit(X_train, y_train)

print("SVM")
evaluate(clf8)
```

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Create base models
base_models = [
    ('rf', AdaBoostClassifier()),
    ('et', GradientBoostingClassifier())
]

# Create meta model (level-2 model)
meta_model = SVC()  # Use SVC from sklearn.svm

# Create stacking classifier
stacking_clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)

# Fit the stacking classifier
stacking_clf.fit(X_train, y_train)

# Evaluate using cross-validation
scores = cross_val_score(stacking_clf, X_train, y_train, cv=5)
print("Stacking Model Accuracy:", scores.mean())

# Make predictions on new data
predictions = stacking_clf.predict(test_df)

# print("Predictions:", predictions)
```
